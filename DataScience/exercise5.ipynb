```python
# Importing Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Breast_Cancer.csv')  # Ensure the file is in the correct path or uploaded to Colab

# Display the first few rows
df.head()
```

```python
sns.countplot(x='diagnosis', data=df, palette={'B':'green', 'M':'red'})
plt.title('Count of Malignant (M) vs Benign (B) Tumors')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()

# Description: This count plot displays how many tumors are malignant (M) and benign (B).
# Insight: Benign cases are more common than malignant in this dataset.
```

```python
corr = df[['radius_mean', 'perimeter_mean', 'area_mean']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Radius, Perimeter, and Area')
plt.show()

# Description: Heatmap of correlation coefficients between size-based features.
# Insight: Very high correlation indicates redundancy — could lead to multicollinearity.
```

```python
corr = df[['radius_mean', 'perimeter_mean', 'area_mean']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Radius, Perimeter, and Area')
plt.show()

# Description: Heatmap of correlation coefficients between size-based features.
# Insight: Very high correlation indicates redundancy — could lead to multicollinearity.
```

```python
for feature in ['concavity_mean', 'smoothness_mean']:
    sns.violinplot(x='diagnosis', y=feature, data=df, palette={'B':'green', 'M':'red'})
    plt.title(f'{feature} by Diagnosis')
    plt.show()

# Description: Violin plots showing feature distributions per class.
# Insight: Malignant tumors have higher concavity; smoothness shows smaller variation.
```

```python
features = [('radius_mean', 'radius_worst'), ('area_mean', 'area_worst')]

for mean_f, worst_f in features:
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df, x=mean_f, hue='diagnosis', kde=True, element='step', palette='Set1')
    plt.title(f'{mean_f} Distribution')

    plt.subplot(1, 2, 2)
    sns.histplot(df, x=worst_f, hue='diagnosis', kde=True, element='step', palette='Set2')
    plt.title(f'{worst_f} Distribution')
    plt.show()

# Description: Side-by-side histograms comparing mean vs worst values.
# Insight: Worst features often provide better class separation than mean values.
```
