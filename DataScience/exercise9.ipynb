Q2. Implement Simple Linear regression based on 
 age = [18, 22, 30, 45, 65, 80]
 accident_no = [38, 36, 24, 20, 18, 28]
(i) Fit and display the regression line for the above data.
(ii) Find number of accidents for the ages 40 and 60.

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Data
car_age = [5,7,8,7,2,17,2,9,4,11,12,9,6]
car_speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

# Reshape data
X = np.array(car_age).reshape(-1, 1)
y = np.array(car_speed)

# Fit model
model = LinearRegression()
model.fit(X, y)

# Predict line
x_line = np.linspace(min(car_age), max(car_age), 100).reshape(-1, 1)
y_line = model.predict(x_line)

# Plot
plt.scatter(car_age, car_speed, color='blue', label='Data points')
plt.plot(x_line, y_line, color='red', label='Regression line')
plt.xlabel('Car Age')
plt.ylabel('Car Speed')
plt.title('Car Age vs Speed')
plt.legend()
plt.grid(True)
plt.show()

# Predict car_speed for car_age = 14
predicted_speed = model.predict(np.array([[14]]))
print(f"Predicted car speed for car age 14: {predicted_speed[0]:.2f}")
```

```python
# Data
age = [18, 22, 30, 45, 65, 80]
accident_no = [38, 36, 24, 20, 18, 28]

# Reshape data
X2 = np.array(age).reshape(-1, 1)
y2 = np.array(accident_no)

# Fit model
model2 = LinearRegression()
model2.fit(X2, y2)

# Predict line
x_line2 = np.linspace(min(age), max(age), 100).reshape(-1, 1)
y_line2 = model2.predict(x_line2)

# Plot
plt.scatter(age, accident_no, color='green', label='Data points')
plt.plot(x_line2, y_line2, color='orange', label='Regression line')
plt.xlabel('Age')
plt.ylabel('Number of Accidents')
plt.title('Age vs Number of Accidents')
plt.legend()
plt.grid(True)
plt.show()

# Predict accidents for age 40 and 60
pred_40 = model2.predict(np.array([[40]]))[0]
pred_60 = model2.predict(np.array([[60]]))[0]
print(f"Predicted number of accidents for age 40: {pred_40:.2f}")
print(f"Predicted number of accidents for age 60: {pred_60:.2f}")
```
